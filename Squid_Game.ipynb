```python
import numpy as np
from scipy.stats import binom
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

n = 18
p = 0.5
rv = binom(n, p)
```

```python
index = np.arange(17)
probs = rv.pmf(index)

from matplotlib.ticker import FuncFormatter
sns.set_theme(style='darkgrid')
ax = sns.lineplot(x=index, y=probs, marker='o')
ax.yaxis.set_major_formatter(FuncFormatter(lambda y, _: '{:.1%}'.format(y))) 
plt.xticks(index)
plt.title('Xác suất có k người chơi chết')
plt.xlabel('k')
plt.ylabel('Xác suất')
```

```python
import matplotlib.pyplot as plt

index = np.arange(16)
probs = rv.cdf(index)
ax = sns.lineplot(x=index, y=probs, marker='o')
ax.yaxis.set_major_formatter(FuncFormatter(lambda y, _: '{:.1%}'.format(y))) 
plt.xticks(index, np.arange(1, 17))
plt.title('Xác suất sống sót của người chơi thứ x')
plt.xlabel('x')
plt.ylabel('Xác suất')
```

```python
pd.DataFrame(data={'Người chơi thứ x': np.arange(1, 17),
                   'Xác suất sống sót': probs})
```
